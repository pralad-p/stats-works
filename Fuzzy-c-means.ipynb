```python
# Common libraries
import numpy as np
from fcmeans import FCM
from matplotlib import pyplot as plt
```

```python
# Make Gaussian distributions
n_samples = 1000

X = np.concatenate((
    np.random.normal((-1, 1), size=(n_samples, 2)),
    np.random.normal((-1, 5), size=(n_samples, 2)),
    np.random.normal((4, 5), size=(n_samples, 2)),
    np.random.normal((4, -1), size=(n_samples, 2)),
))

plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
# Apply F-C-Means algo.
fcm = FCM(n_clusters=4)
fcm.fit(X)
```

```python
# outputs
fcm_centers = fcm.centers
fcm_labels = fcm.predict(X)

# plot result
f, axes = plt.subplots(1, 2, figsize=(11,5))
axes[0].scatter(X[:,0], X[:,1], alpha=.3)
axes[1].scatter(X[:,0], X[:,1], c=fcm_labels, alpha=.5)
axes[1].scatter(fcm_centers[:,0], fcm_centers[:,1], marker="+", s=500, c='w')
plt.show()
```
